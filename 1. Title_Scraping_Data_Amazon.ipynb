The User-Agent HTTP Header is a header that contains a string that tells the web server what kind of operating system and browser it is.

whatismybrowser.com --> Detect My Settings --> What is my user agent

```python
URL = "https://www.amazon.in/s?k=watches&crid=4LI5NUNBDFHQ&sprefix=watches%2Caps%2C280&ref=nb_sb_noss_1"
```

```python
# Headers for request
HEADERS = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent 
```

```python
# HTTP Request
webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
# Soup Object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

Beautiful Soup is a Python library that is used for web scraping purposes to pull the data out of HTML and XML files. 

```python
# Soup Object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
soup.prettify()
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
```

```python
links
```

```python
link = links[0].get('href')
link
```

```python
product_list = "https://amazon.in" + link
product_list
```
